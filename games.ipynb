## Подготовка данных

### Приведение столбцов к нижнему регистру

```python
df.columns = df.columns.str.lower()
df.columns
```

Необходимо заменить типы данных в следующих столбцах:
* `year_of_release` - год - целое число, приведем к типу `int`;
* `critic_score` - оценка критиков - увидели, что в датасете выставляются только целые числа, приведем к типу `int`;
* `user_score` - оценка пользователей - десятичная дробь, приведем к типу `float`.

Но сначала обработаем пропуски, которые помешают в изменении типа данных

### Обработка пропусков

```python
df.isna().sum()
```

```python
df[df['name'].isna()]
```

```python
df['user_score'].value_counts()
```

```python
df['critic_score'].unique()
```

Возможные причинны возникновения пропусков:
* пропуски в `name`совпадают с пропусками в `genre`, эти 2 строки малоинформативны и их удаление не повлияет на результат анализа;
* пропуски в `year_of_release` - данные о годе выпуска игры могли не подгрузиться, строк не много, но в них есть данные по продажам, игры могли выходить в разные годы, в том числе на разных платформах, заменим на -1;
* пропуски в `critic_score`, `user_score`- возможно, к этим играм еще не были добавлены оценки на момент выгрузки данных или эти игры не оценивались. По мимо пропусков, в столбце `user_score` есть значение 'tbd', что означает "To Be Decided" или "To Be Determined" ("будет определено/решено") - скорее всего, недостаточно оценок для определения рейтинга. Данных с пропусками в этих столбцах много и их необходимо сохранить. Для перевода к нужному типу пропуски в `critic_score` заменим на -1. Пропуски и 'tbd' в `user_score` объединим в одну группу, так как в обоих случаях рейтинга нет, заменим на -1;
* пропуски в `rating` - возможно ассоциация не определила возрастной рейтинг игры, или данные не подгрузились, заменим на 'no_rating'

```python
df = df.drop([659,14244])
```

```python
df['rating'] = df['rating'].fillna('no_rating')
df['critic_score'] = df['critic_score'].fillna(-1)
df['user_score'] = df['user_score'].fillna(-1)
df['year_of_release'] = df['year_of_release'].fillna(-1)
df.loc[(df['user_score'] == 'tbd'), 'user_score'] = -1
```

### Распределение продаж по платформам

```python
#Сгруппируем игры по платформе, посчитаем суммарную выручку платформы, построим график
platform_sales = pd.pivot_table(df1, index='platform', values='all_sales', aggfunc='sum')
platform_sales = platform_sales.sort_values(by='all_sales', ascending=False)
platform_sales.plot(kind='barh', legend=False, figsize=(7,7))
plt.title('Суммарные продажи игр')
plt.xlabel('Количество проданных копий, млн')
plt.ylabel('Платформа');
```

Лидирующую позицию по количеству продаж копий игры занимает платформа PS2

```python
#Отобразим на графике распределение количества проданных копий 5 наиболее популярных платформ из имеющихся в датасете
for i in (platform_sales[:5].reset_index())['platform'].unique():
    df1.query('platform == @i')\
    .pivot_table(index = 'year_of_release',
                 values = 'all_sales', 
                 aggfunc = 'sum').plot(kind = 'bar', grid=True, legend=False, figsize=(8,5))
    plt.title('Суммарные продажи игр на платформе' + ' ' + i)
    plt.xlabel('Год')
    plt.ylabel('Количество проданных копий, млн')
```

Наиболее прибыльные платформы за весь период:
* PS2 - приносила прибыль с 2000 по 2011 год, продажи стали падать с 2005 года;
* X360 - приносила прибыль с 2005 по 2016 год, продажи стали падать с 2011 года;
* PS3 - приносила прибыль с 2006 по 2016 год, продажи стали падать с 2013 года;
* Wii - приносила прибыль с 2006 по 2016 год, продажи стали падать с 2010 года;

### Влияние отзывов критиков и пользователей на продажи игр

```python
platform_list = ['PS3', 'X360', 'PS4', '3DS', 'XOne']
for platform in platform_list: 
  plat = df_actual.query('platform == @platform and critic_score != -1')
  plat.plot (x='critic_score', y ='all_sales', kind='scatter', alpha=0.4, figsize=(7,4))
  plt.title(f'Зависимость продаж игр на платформе {platform} от оценки критиков')
  plt.xlabel('Оценка критиков')
  plt.ylabel('Количество проданных копий, млн')
  plt.show()
  print(f'Коэффициент корреляции оценки критиков и продаж игр на платформе {platform}:', plat['all_sales'].corr(plat['critic_score']).round(3))
```

```python
platform_list = ['PS3', 'X360', 'PS4', '3DS', 'XOne']
for platform in platform_list: 
  plat = df_actual.query('platform == @platform and user_score != -1')
  plat.plot(x='user_score', y ='all_sales', kind='scatter', alpha=0.4, figsize=(7,4))
  plt.title(f'Зависимость продаж игр на платформе {platform} от оценки пользователей')
  plt.xlabel('Оценка пользователей')
  plt.ylabel('Количество проданных копий, млн')
  plt.show()
  print(f'Коэффициент корреляции оценки пользователей и продаж игр на платформе {platform}:', plat['all_sales'].corr(plat['user_score']).round(3))
```

## Проверка гипотез

Нулевая гипотеза формулируется так, что по умолчанию предполагается, что никаких различий в двух выборках нет

### Проверка гипотезы: средние пользовательские рейтинги платформ Xbox One и PC одинаковые


Нулевая гипотеза: средние пользовательские рейтинги платформы Xbox One = средние пользовательские рейтинги платформы PC

Альтернативная гипотеза: средние пользовательские рейтинги платформы Xbox One ≠ средние пользовательские рейтинги платформы PC

Установим уровень значимости alpha = 0.05

```python
xone = df_actual[df_actual['platform'] == 'XOne']
xone = xone.query('user_score != -1')
xone.shape
```

```python
pc = df_actual[df_actual['platform'] == 'PC']
pc = pc.query('user_score != -1')
pc.shape
```

Необходимо проверить гипотезу о равенстве среднего двух генеральных совокупностей по взятым из них выборкам, выборки не зависимы, поэтому будем использовать `scipy.stats.ttest_ind`. Выборки разного размера, поэтому укажем параметр `equal_var=False`

```python
results = st.ttest_ind(xone['user_score'], pc['user_score'], equal_var=False)
alpha = 0.05
print(results.pvalue)
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

```python
#посмотрим распределение оценок
xone.hist('user_score', grid=True, legend=False, figsize=(7,5))
plt.title('Распределение оценки пользователей платформы XOne')
plt.xlabel('Оценка')
plt.ylabel('Количество оценок')
plt.show()
pc.hist('user_score', grid=True, legend=False, figsize=(7,5))
plt.title('Распределение оценки пользователей платформы PC')
plt.xlabel('Оценка')
plt.ylabel('Количество оценок')
plt.show();
```

Значит принимаем нулевую гипотезу о том, что средние пользовательские рейтинги платформ Xbox One и PC одинаковые

```python
xone['user_score'].hist(bins=10, alpha=0.5, color='green',figsize=(7,5))
pc['user_score'].hist(bins=10, alpha=0.5, color='blue', figsize=(7,5))
plt.title('Распределение оценки пользователей платформы XOne и PC');
```

### Проверка гипотезы: средние пользовательские рейтинги жанров Action и Sports разные

Нулевая гипотеза: средние пользовательские рейтинги жанров Action = средние пользовательские рейтинги жанров Sports

Альтернативная гипотеза: средние пользовательские рейтинги жанров Action ≠ средние пользовательские рейтинги жанров Sports

Установим уровень значимости alpha = 0.05

```python
action = df_actual[df_actual['genre'] == 'Action']
action = action.query('user_score != -1')
action.shape
```

```python
sports = df_actual[df_actual['genre'] == 'Sports']
sports = sports.query('user_score != -1')
sports.shape
```

```python
results = st.ttest_ind(action['user_score'], sports['user_score'], equal_var=False)
alpha = 0.05
print(results.pvalue)
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

```python
#посмотрим распределение оценок
action.hist('user_score', grid=True, legend=False, figsize=(7,5))
plt.title('Распределение оценки пользователей жанра Action')
plt.xlabel('Оценка')
plt.ylabel('Количество оценок')
plt.show()
sports.hist('user_score', grid=True, legend=False, figsize=(7,5))
plt.title('Распределение оценки пользователей жанра Sports')
plt.xlabel('Оценка')
plt.ylabel('Количество оценок')
plt.show();
```

```python
action['user_score'].hist(bins=10, alpha=0.5, color='green',figsize=(7,5))
sports['user_score'].hist(bins=10, alpha=0.5, color='blue', figsize=(7,5))
plt.title('Распределение оценки пользователей жанра Action и Sports');
```

Значит принимаем альтернативную гипотезу о том, что средние пользовательские рейтинги жанров Action и Sports разные
